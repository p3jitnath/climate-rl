```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def plot_images_in_grid(image_paths, grid_dim, output_path=None):
    fig, axes = plt.subplots(
        nrows=grid_dim[0], ncols=grid_dim[1], figsize=(16, 16)
    )

    for idx, ax in enumerate(axes.flatten()):
        if idx < len(image_paths):
            img = Image.open(image_paths[idx])
            img = np.array(img)
            ax.imshow(img)
            ax.axis("off")
        else:
            ax.axis("off")

    plt.subplots_adjust(
        wspace=0.0, hspace=0.0
    )  # Adjust spacing between images
    if output_path:
        plt.savefig(output_path, dpi=300, bbox_inches="tight")
    else:
        plt.show()
```

```python
BASE_DIR = "/gws/nopw/j04/ai4er/users/pn341/climate-rl"
IMGS_DIR = f"{BASE_DIR}/results/imgs/"
```

```python
exp_ids = [
    "scbc-v0-optim-L",
    "scbc-v1-optim-L",
    "scbc-v2-optim-L",
    "scbc-v0-optim-L-60k",
    "scbc-v1-optim-L-60k",
    "scbc-v2-optim-L-60k",
    "scbc-v0-homo-64L",
    "scbc-v1-homo-64L",
    "scbc-v2-homo-64L",
    "scbc-v0-homo-64L-60k",
    "scbc-v1-homo-64L-60k",
    "scbc-v2-homo-64L-60k",
]

img_paths = [
    IMGS_DIR + x + "_log10_top3_episodic_returns.png" for x in exp_ids
]

output_path = f"{BASE_DIR}/results/imgs/scbc_x12_v012_merged.png"

plot_images_in_grid(img_paths, (4, 3), output_path)
```
