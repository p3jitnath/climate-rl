```python
BASE_DIR = "/gws/nopw/j04/ai4er/users/pn341/climate-rl"
RUNS_DIR = f"{BASE_DIR}/runs"
TABLES_DIR = f"{BASE_DIR}/results/tables/"

exp_id = exp_ids[0]
df = pd.read_csv(f"{TABLES_DIR}{exp_id}.csv")
df = df.dropna()
df["steps_to_threshold"] = df["steps_to_threshold"].astype(int)
df = df.drop(["score"], axis=1)
df
```

```python
import re


def generate_latex_string(df, exp_id):
    add_string = (
        "\\begin{subtable}{\\textwidth}\n\centering\n"
        + f"\caption{{\\texttt{{{exp_id}}}}}\n\label{{tab:app-{exp_id}}}\n"
    )
    latex_string = re.sub(
        "<([^>]*)>",
        "\\\\textbf{\g<1>}",
        df.rename(columns=lambda x: f"<{x}>").to_latex(
            index=False, escape=True
        ),
    )

    return add_string + latex_string + "\end{subtable}"


print(generate_latex_string(df, exp_id))
```

```python
for exp_id in exp_ids:
    df = pd.read_csv(f"{TABLES_DIR}{exp_id}.csv")
    df = df.dropna()
    df["steps_to_threshold"] = df["steps_to_threshold"].astype(int)
    df = df.drop(["score"], axis=1)
    print(generate_latex_string(df, exp_id))
    print("\n")
```
